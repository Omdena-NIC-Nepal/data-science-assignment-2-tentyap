Basic of ploting

```python
# Line ploting
x = [1,2,4,5,6,7]
y = [7,8,9,11,2,15]
```

```python
plt.plot(x,y)
plt.xlabel("x -axis")
plt.ylabel("y -axis")
plt.title("simple line Chart") 
plt.show() 
```

```python
#Scatter Plot 
plt.scatter(x,y)
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title(" simple scatter plot")
plt.show()
```

```python
## Bar plot

## compare of Quantites

catogory = ['A','B','C','D']
values = [2,4,15,11]

```

```python
plt.bar(catogory,values)
plt.xlabel('catogory')
plt.ylabel('values')
plt.title("Bar digram")
plt.show()

```

```python
data = [1,2,3,4,5,6,7,8,9,22,34,2,4,22,22,22]
plt.hist(data, bins =100 ) # bins show the zoom in and zoom out of the line
plt.xlabel('values')
plt.ylabel('frequency')# it show how many time does the number repeated if data is reapeated more time it will show higher
plt.title('histogram')
plt.show()
```

# Basic ploting of the seaborn


```python
# scatter plot 

data = pd.DataFrame({
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eva', 'Frank', 'Grace'],
    'Age': [25, 45, 30, 50, 35, 40, 28],
    'Email': ['alice@example.com', 'bob@example.com', 'charlie@example.com', 'david@example.com', 'eva@example.com', 'frank@example.com', 'grace@example.com'],
    'Phone': ['555-1234', '555-5678', '555-9101', '555-1121', '555-3141', '555-5161', '555-7181'],
    'Salary': [55000, 80000, 60000, 120000, 75000, 95000, 70000],
    'Height': [165, 180, 175, 170, 160, 185, 168],
    'Net Worth': [150000, 250000, 200000, 500000, 300000, 400000, 220000]
})
```

```python
print(data)
```

```python
# scatter plot 
sns.scatterplot(x ="Name", y="Net Worth", data=data)
plt.title('scatter plot')
plt.show()
```

```python
# Line plot 
sns.lineplot(x="Name",y="Height", data=data)
plt.title("Height of the person")
plt.show()
```

```python
data.info()
```

```python
# Bar plot 
sns.barplot(x= "Name", y="Net Worth", data=data )
plt.title("Bar digram of person Accroding to Height")
plt.show()
```
